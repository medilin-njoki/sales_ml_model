```python
def drop_closed_rows(df, open_col="Open"):
    return df[df[open_col] != 0].copy()
```

```python
df_open = drop_closed_rows(df)
df_open.head()
```

```python
df_open["Date"] = pd.to_datetime(df_open["Date"])

df_open["Year"] = df_open["Date"].dt.year
df_open["Month"] = df_open["Date"].dt.month
df_open["Day"] = df_open["Date"].dt.day
```

```python
df_open
```

```python
def date_features(df, date_col="Date"):
    df = df.copy()
    df[date_col] = pd.to_datetime(df[date_col])
    df["Year"] = df[date_col].dt.year
    df["Month"] = df[date_col].dt.month
    df["Day"] = df[date_col].dt.day
    return df
```

```python
df_open = date_features(df_open)
df_open.head()
```

```python
df_open = df_open.drop(columns=["Date", "True_index"])
```
